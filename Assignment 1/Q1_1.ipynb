````python
class Node:
    def __init__(self, threshold=None, left=None, right=None, feature_index=None, value=None):
        self.threshold = threshold
        self.value = value
        self.left = left
        self.right = right
        self.feature_index = feature_index

    def __repr__(self):
        return f"Node(threshold={self.threshold}, value={self.value}, left={self.left}, right={self.right}, feature_index={self.feature_index})"

    def isLeaf(self):
        return (self.value != None)
````

````python
#@title Decision Tree Class

class DecisionTree:
    def __init__(self, min_samples = 2, max_depth=15, impurity="entropy"):
        self.max_depth = max_depth
        self.min_samples = min_samples
        self.root = None
        self.impurity = impurity # impurity function


    def build_tree(self, data, depth):
        num_samples , num_features = np.shape(data[:,:-1])
        if num_samples >= self.min_samples and depth <= self.max_depth and num_samples != 1: # Done for last experiment
          split = self.get_best_split(data)
          if split['info_gain'] > 0:
            # valid split, recurse itno left and right
            lchild, rchild = self.build_tree(split['lchild'], depth + 1), self.build_tree(split['rchild'], depth + 1)
            return Node(split['threshold'], lchild, rchild, split['feature_index'] )

        # leaf node
        value = self.calculate_leaf_value(data)
        return Node(value=value)



    def get_best_split(self, data):
        num_samples , num_features = np.shape(data[:,:-1])
        max_info_gain = -float("inf")
        best_split = {}
        for i in range(num_features):
          feature_val = data[:, i]
          unique_thresholds = np.unique(feature_val)
          for threshold in unique_thresholds:
            df_left = np.array([x for x in data if x[i] <= threshold])
            df_right = np.array([x for x in data if x[i] > threshold])
            if len(df_left) == 0 or len(df_right) == 0: continue
            original, left, right = data[:, -1], df_left[:, -1], df_right[:,-1]
            information_gain = self.get_information_gain(original, left, right)

            if information_gain > max_info_gain:
              max_info_gain = information_gain
              self.set_info_gain(i, df_left, df_right, threshold, max_info_gain, best_split)

        return best_split

    def set_info_gain(self, i, df_left, df_right, threshold, information_gain, best_split):
        best_split.update({
          "feature_index": i,
          "lchild": df_left,
          "rchild": df_right,
          "threshold": threshold,
          "info_gain": information_gain
        })
        return

    def get_information_gain(self, original, left, right):
        wt_lchild, wt_rchild = len(left) / len(original) , len(right)/len(original)
        if self.impurity == "gini":
          IG = 1 * self.gini_impurity(original) - (wt_lchild * self.gini_impurity(left) + wt_rchild*self.gini_impurity(right))

        else:
          IG = 1 * self.entropy(original) - (wt_lchild * self.entropy(left) + wt_rchild*self.entropy(right))

        return IG

    def entropy(self, data):
        class_labels, class_counts = np.unique(data, return_counts=True) # gets all the unique labels, the bool is for getting sum directly
        class_probs = class_counts / len(data) # probability of each class
        entropy = -np.sum(class_probs * np.log2(class_probs))
        return entropy

    def gini_impurity(self, data):
      class_labels, class_counts = np.unique(data, return_counts=True)
      class_probs = class_counts / len(data)
      gini_impurity = 1 - np.sum(class_probs**2)
      return gini_impurity

    def calculate_leaf_value(self, data):
        y = data[:, -1]
        Y = list(y)
        mode_value = None
        max_count = 0

        for value in set(Y):
            count = Y.count(value)

            if count > max_count:
                max_count = count
                mode_value = value

        return mode_value

    def fit(self, data):
      self.root = self.build_tree(data, 0)

    def predict(self, data):

        predictions = [self.make_prediction(x, self.root) for x in data]
        return predictions

    def make_prediction(self, x, node):

        if node.isLeaf(): return node.value
        feature_val = x[node.feature_index]
        if feature_val<=node.threshold:
            return self.make_prediction(x, node.left)
        else:
            return self.make_prediction(x, node.right)
````

````python
X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values.reshape(-1,1)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.8, random_state=42)
````

````python
from sklearn.metrics import accuracy_score
import time

acc_entropy = []
time_taken_entropy = []

for prune in range(1, 16):
    model = DecisionTree(min_samples=3, max_depth=prune)
    dataset = np.concatenate((X_train, Y_train), axis=1)

    start_time = time.time()

    model.fit(dataset)
    Y_pred = model.predict(X_test)

    end_time = time.time()

    curr_acc = accuracy_score(Y_test, Y_pred)
    acc_entropy.append(curr_acc)
    time_taken_entropy.append((tt := (end_time - start_time)))
    print(f"time for prune: {prune} : {tt}, accuracy: {curr_acc}")

print("accuracy:", acc_entropy)
print("time: ", time_taken_entropy)
````

## Experiment:
If we remove the min_samples argument, rather than re-writing an entire class, we just set it to the ```minimum value = 1``` since it requires atleast 2 features to split a node further, we are expecting some over-fitting since the tree would now contain leaf nodes with just one sample and try to predict. Not sure about the time it would take, estimating it would take more time since we split further now.

**Note (1.1?)**: For this I have added the condition in the tree (num_samples > 1) since we dont want to split it further because we would obviously get either lchild/rchild as `None` type
